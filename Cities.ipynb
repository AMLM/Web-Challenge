```python
import pandas as pd
import numpy as np
```

```python
csv_path = "Resources/cities.csv"
cities_df = pd.read_csv(csv_path)
```

```python
cities_df.head()
```

```python
datesInt = []
for index, row in cities_df.iterrows():
    datesInt.append(int(row[3]))

cities_df['DatesInt'] = datesInt
```

```python
cities_df.head()
```

Converting dataframe to HTML for visualization in webpages.

```python
cities_visual = cities_df[['City', 'Cloudiness', 'Country', 'Humidity', 'Lat', 'Lng', 'Max Temp', 'Wind Speed', 'DatesInt']]
```

```python
cities_visual.to_html("cities_visual.html")
```
